

**Perform histogram equalization of a gray scale image.**

```python
import cv2
import matplotlib.pyplot as plt

# Read the grayscale image
img_gray = cv2.imread("image.tif", cv2.IMREAD_GRAYSCALE)

# Perform histogram equalization
equalized_img = cv2.equalizeHist(img_gray)

# Display the original and equalized images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_gray, cmap='gray')
plt.title('Original Grayscale Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(equalized_img, cmap='gray')
plt.title('Histogram Equalized Image')
plt.axis('off')

plt.show()
```

**User Perform image smoothing using Gaussian filtering.**

```python
import cv2
import matplotlib.pyplot as plt

# Read the image
img = cv2.imread("image.tif")

# Convert the image to grayscale (optional)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur
# You can adjust the kernel size (sigmaX and sigmaY) based on your requirement
# Larger kernel size will result in more blurring
blurred_img = cv2.GaussianBlur(img_gray, (5, 5), sigmaX=0)

# Display the original and blurred images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(blurred_img, cmap='gray')
plt.title('Gaussian Blurred Image')
plt.axis('off')

plt.show()
```

**Apply median filtering to remove noise.**

```python
import cv2
import matplotlib.pyplot as plt

# Read the image
img = cv2.imread("image.tif")

# Apply median filtering
# You can adjust the kernel size based on the level of noise
# Larger kernel sizes will result in stronger noise reduction, but may also blur the image
median_filtered_img = cv2.medianBlur(img, 5)

# Display the original and median filtered images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(median_filtered_img, cv2.COLOR_BGR2RGB))
plt.title('Median Filtered Image')
plt.axis('off')

plt.show()
```
